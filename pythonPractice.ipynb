```python
"""live_coding_amazon"""

import datetime


class Customer:
    def __init__(self, name, email, products):
        self.name = name
        self.email = email
        self.products = products
        self.products_ordered = {}

    def order_item(self, name):
        product = self.products.find_item_by_name(name)
        if product:
            self.products_ordered[product] = datetime.datetime.today()
        else:
            print("Could not find that product")

    def print_ordered_items(self):
        for key, value in self.products_ordered.items():
            print(key.name, key.price)


class Product:
    def __init__(self, name, price, shipping_days):
        self.name = name
        self.price = price
        self.in_stock = True
        self.shipping_time = shipping_days


class ProductList:
    def __init__(self):
        self.products = []

    def create_product(self, name, price, shipping_days):
        new_product = Product(name, price, shipping_days)
        self.products.append(new_product)

    def find_item_by_name(self, name):
        for product in self.products:
            if product.name == name:
                return product


products = ProductList()
rick = Customer('Rick', "rick@gmail.com", products)
products.create_product("Google Home", 49.99, 3)
products.create_product("Apple Watch", 300, 5)
products.create_product("xbox", 200, 7)
rick.order_item("Google Home")
rick.order_item("Apple Watch")
rick.print_ordered_items()
```

```python
"""
Data Structure Exercise

Stack: LIFO

Queue: FIFO
"""

class Stack:
    def __init__(self):
        self.items = []

    def is_empty(self):
        return self.items == []

    def push(self, item):
        self.items.append(item)

    def pop(self):
        return self.items.pop()

    def peek(self):
        last = len(self.items)-1
        return self.items[last]

    def size(self):
        return len(self.items)
    
class Queue:
   def __init__(self):
       self.items = []


   def is_empty(self):
       return self.items == []


   def enqueue(self, item):
       self.items.insert(0, item)


   def dequeue(self):
       return self.items.pop()


   def size(self):
       return len(self.items)


#============Problem 1=============
stack=Stack()

for i in "yesterday":
    stack.push(i)
print(stack.items)

reverse_str = ""
for i in range(stack.size()):
    reverse_str+=stack.pop()
print(reverse_str)

#============Problem 2=============
stack2=Stack()

ls = [1, 2, 3, 4, 5]
rev_ls = []

for i in ls:
    stack2.push(i)
print(stack2.items)

for i in range(stack2.size()):
    rev_ls.append(stack2.pop())
print(rev_ls)
```

```python
"""
Recursion

A method to solve a problem by breaking it down 
to smaller pieces that are much easier to solve.

In programming, recursive algorithm relies on the repetition 
of a function within itself until the algorithm satisfies the 
base case. With that being said, a recursive algorithm 
is often written inside a function. The function will 
then call itself until it meets the base csse.

Base case is the condition(s) set to end the recursive 
algorithm to prevent it from running forever.

Three laws of recursion:
    1. must have a base case
    2. must change its state and move toward the base case
    3. must repeat itself 

"""

#============Example=============
def bottles_of_beer(bob):
    if bob < 1:
        print("""No more bottles of beer on the wall. No more 
                 bottles of beer.""")
        return
    tmp = bob
    bob -= 1
    print("""{} bottles of beer on the wall. {} bottles
             of beer. Take one down, pass it around, 
             {} bottles of beer on the wall.""".format(tmp, tmp, bob))
    bottles_of_beer(bob)
    
#============Problem=============
"My recursive function: Lyrics to the Five little ducks"

def five_lil_ducks(left_duck=5, back_duck=0):
    print('\nSad mother duck went out one day.\
        \nOver the hill and far away.\
        \nThe sad mother duck said “Quack, quack, quack,”')
    if back_duck==0:
        print("And none of those little ducks came back.")
    if back_duck!=0 and back_duck!=left_duck:
        print("And only {} little duck(s) came back."\
         .format(back_duck))
    if left_duck==back_duck:
        print("And all of the {} little ducks came back."\
             .format(back_duck))
        return 
    back_duck+=1
    five_lil_ducks(left_duck, back_duck)
        
five_lil_ducks(4)
```

```python
"""
find_unique.py

Author: Ye N.E.

Description: Given a non-empty list of integers, 
            every element appears twice except for one. 
            Find that single one.
"""

def find_unique(ls):
    dup=[]
    a_set=set()
    for i in ls:
        length_1=len(a_set)
        a_set.add(i)
        length_2=len(a_set)
        if length_1==length_2:
            dup.append(i)
    for j in ls:
        if j not in dup:
            return j
        
ls = [1 , 2, 3, 1, 3, 4, 5, 1, 4, 5]
find_unique(ls)
    
```
